**Defining Helper Functions**

```python
# The following sequence of functions handle the conversion to name list  

class NameConverter():
    
        

    def strip_sentinels(self, col_value):

        sentinels = ['[',']']

        for char in sentinels:

            col_value = col_value.strip(char)

        return col_value


    def format_col_value(self, clean_string):

        temp_list = clean_string.split('}')

        name_list = []

        for val in temp_list[:-1]:    # temp_list[:-1] since last element is null string.

            val = val + '}'     # The '}' character was stripped off when the split was carried on based off it. 

            val = val.strip(',')  # Stripping off the comma present in second and subsequent elements.

            val = eval(val)    # Converting each value to the dictionary datatype.

            val = val.get('name')  

            name_list.append(val)  

        return name_list


    def main(self,x):
        
        name_c = NameConverter()
        
        cleaned_string = name_c.strip_sentinels(x)

        transformed_value =  name_c.format_col_value(cleaned_string)

        return transformed_value

    
```

**Transformations**

#### 1 . Transforming column values

```python
# Transforming column values to name list

# Instantiating converter class

name_converter = NameConverter()

# Executing Transformations

movies.genres = movies.genres.apply(name_converter.main)
movies.spoken_languages =  movies.spoken_languages.apply(name_converter.main)
movies.production_companies = movies.production_companies.apply(name_converter.main)
movies.production_countries = movies.production_countries.apply(name_converter.main)
movies.keywords = movies.keywords.apply(name_converter.main)
```

```python
# Rounding of columns values for popularity

movies.popularity =  movies.popularity.apply(np.round, args = [4])
```

#### 2. Transforming columns to appropriate type

```python
# Converting release_date to datetime

movies.release_date = pd.to_datetime(movies.release_date)   
```

#### 3. Creating new columns

```python
# Creating column: profit

movies['gross_profit'] = movies.revenue - movies.budget
```

#### 4. Handling Null Values

```python
movies.drop(columns = ['homepage','tagline'], inplace = True) 
```

## Exploratory Data Analysis


### Research Question 1: What genres are the most popular from year to year?

**Approach** 

To find the most popular genres from year to year - we'll be creating a dataframe by extracting the individual genre for 
each movie with their corresponding popularity score and the year of release. After which, we'll group the data by release year and aggregate it by the maximum popularity column thus yielding the most popular genre for each year.

This method captures the rating each genre recieved which helps us in ranking them for their popularity
for a particular year.


Moreover, we'll then calculate the most popular genres by counting the number of times each genre featured as the most 
popular genre for a year. For this, we'll group the data based on genre and aggregate it using the count function.

**Most Popular Genre by Year**

```python
Genres = []
Popularity = []
Release_date = []

for genres, popularity, release_dt in zip(movies.genres, movies.popularity, movies.release_date):
    
    for genre in genres:
        
        Genres.append(genre)
        Popularity.append(np.round(popularity,3))
        Release_date.append(release_dt.year)


genre_popularity = pd.DataFrame({'Genre' : Genres,
                                 'Popularity': Popularity,
                                 'Release_date' : Release_date
                                })

gen_p = genre_popularity.groupby(genre_popularity.Release_date).max().sort_values(by = 'Release_date', ascending = False)

pd.options.display.max_columns = 90
gen_p.unstack().unstack()
```

**Most Popular Genres in Aggregate**

```python
gp = gen_p.groupby(gen_p.Genre).count().sort_values(by = 'Popularity', ascending = False)
gp.rename( columns = {'Popularity' : 'Popular Count'}, inplace = True)

gp
```

**Visualization**

We will be using a horizontal bar chart to visualize the performance of genres. Such a chart helps us to see the magnitude of difference in popularity of genres.

```python
# Visualizing the ranking of genres

G = sns.barplot(y = gp.index, x = gp['Popular Count'])
sns.despine()

G.set_title('Best Performing Genres', pad = 15)
G.set_xlabel('Popular Count', fontdict = {'size': 12})
G.set_ylabel('Genre', fontdict = {'size': 12})
```

** **

**Significant Observations**:

1. The most popular genres in aggregate is Western, followed by War, and then by Romance & Thriller. 

2. The most popular genre Western: also happens to feature as one for a number of years in succesion.

### Research Question 2: What kinds of properties are associated with movies that have high revenues?

**Approach**

To find the properties that are associated with movies having higher revenues, we'll find how the various 
numeric variables are correlated with revenues and take note of all variables having a correlational coefficient
of higher than 0.5. We are doing this to discover non-intuitive associations.


Note: This approach only accounts for the numeric variables as determining the relationships with categorical variables 
would require some advanced statistical techniques. 

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.

feature_cor_revenue = pd.DataFrame(movies.select_dtypes(include = np.number).drop(columns = ['id','revenue']).\
                                                                 corrwith(movies.revenue).\
                                                                 sort_values(ascending = False),
                                                                 columns = ['Correlation coefficient with revenue']
                                                                  )

feature_cor_revenue
```

**Visualization**

We'll be using seaborn's regplot to depict the linear associations of variables. The use of regplot enables us to 
see the line of best fit and intepret the strength of these variables.

```python
# Scatterplots to depict the association of variables

fig = plt.figure()
fig.suptitle('Features Associated with Revenue')


fig.add_subplot(2,2,1)
sns.regplot(y = movies.gross_profit, x = movies.revenue)

fig.add_subplot(2,2,2)
sns.regplot(y = movies.vote_count, x = movies.revenue)

fig.add_subplot(2,2,3)
sns.regplot(y = movies.budget, x = movies.revenue)

fig.add_subplot(2,2,4)
sns.regplot(y = movies.popularity, x = movies.revenue)

plt.subplots_adjust(wspace = 0.6, hspace = 0.6)
```

**Findings**

* *Gross profit*, *vote count*, *budget*, and *popularity* are the factors that are associated with higher revenues. 

1. Since gross profit is a variable derieved from revenue - it was expected to have a strong linear relationship with it.


2. Vote count is the number of participants involved in rating the film and a higher correlation with revenue implies
higher number of voters are associated with higher revenues.


3. A high correlation of budget with revenues implies that higher budget films are strongly associated with higher revenues. 


4. Finally, a high popularity rating is associated with higher revenues which was expected.


To sum up: The major variables associated with higher revenues are **vote_count**, **budget**, & **popularity** in that order.
